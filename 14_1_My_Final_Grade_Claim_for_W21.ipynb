## NP-Completeness

During Week 12's [BDD CDL]( https://colab.research.google.com/github/croesusking/cse480-notebooks/blob/master/12_3_More_About_Boolean_Logic_and_BDDs.ipynb), I share my insight to my team mates about why Boolean reasoning is in P, while Boolean satisfiability is in NP.
I remarked that Boolean satisfiability would check if there is at least one combination of equations that evaluates to true. This must be NP-Complete as in the worst case you would have to check all  2n  (where n is the number of variables) variable combinations to determine if the Boolean statement is satisfiable. It must be in NP because the algorithm to check a potential answer to Boolean satisfiability is just Boolean reasoning, which is in P.

From this - we went on to complete the Ponder and Prove for Binary Decision Diagrams - [BDD P&P](https://colab.research.google.com/github/croesusking/cse480-notebooks/blob/master/12_1_Ponder_and_Prove_Binary_Decision_Diagrams.ipynb). We noted how the order of the optimal variable ordering would be to start from the most significant bit to the least significant bit since the reverse would require us to traverse the whole path as opposed to starting from the most significant bit.
